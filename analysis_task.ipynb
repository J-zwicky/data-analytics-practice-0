# TASK : highest number of PM10 stations per 1000sqkm.

```python
import pandas as pd
import requests
import geopandas as gpd
from io import StringIO

# Fetch the GeoJSON data
url = "https://api.energyandcleanair.org/stations?country=GB,US,TR,PH,IN,TH&format=geojson"
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content to a file-like object for geopandas
    geojson_data = StringIO(response.text)
    stations = gpd.read_file(geojson_data)  # Read GeoJSON into a GeoDataFrame
    
    # Print the first few rows
    print(stations.head())
else:
    print(f"Failed to fetch data: {response.status_code}")
```

```python
stations_df = stations
stations_df.head()
```

```python
# Fetch the GeoJSON data
url_1 = "https://r2.datahub.io/clvyjaryy0000la0cxieg4o8o/main/raw/data/countries.geojson"
response = requests.get(url_1)

# Check if the request was successful
if response.status_code == 200:
    # Save the content to a file-like object for geopandas
    geojson_data_1 = StringIO(response.text)
    country_boundaries = gpd.read_file(geojson_data_1)  # Read GeoJSON into a GeoDataFrame
    
    # Print the first few rows
    print(country_boundaries.head())
else:
    print(f"Failed to fetch data: {response.status_code}")
```

```python
country_boundaries_df = country_boundaries
country_boundaries_df
```

```python
country_boundaries_df.loc[country_boundaries_df['name'] == "Thailand"]
```

```python
TH_boundry = country_boundaries_df.loc[country_boundaries_df['name'] == "Thailand"]
TR_boundry = country_boundaries_df.loc[country_boundaries_df['name'] == "Turkey"]
IN_boundry = country_boundaries_df.loc[country_boundaries_df['name'] == "India"]
PH_boundry = country_boundaries_df.loc[country_boundaries_df['name'] == "Philippines"]
GB_boundry = country_boundaries_df.loc[country_boundaries_df['ISO3166-1-Alpha-3'] == "GBR"]
US_boundry = country_boundaries_df.loc[country_boundaries_df['name'] == "United States of America"]
required_boundaries_df = pd.concat([TH_boundry,TR_boundry,GB_boundry,PH_boundry,IN_boundry,US_boundry])
```

```python
count_US = stations_df['country_id'].value_counts().get('US',0)
count_TR = stations_df['country_id'].value_counts().get('TR',0)
count_GB = stations_df['country_id'].value_counts().get('GB',0)
count_TH = stations_df['country_id'].value_counts().get('TH',0)
count_PH = stations_df['country_id'].value_counts().get('PH',0)
count_IN = stations_df['country_id'].value_counts().get('IN',0)
print("Stations in USA :",count_US)
print("Stations in TURKEY :",count_TR)
print("Stations in GREAT BRITAIN :",count_GB)
print("Stations in THAILAND :",count_TH)
print("Stations in PHILIPPINES :",count_PH)
print("Stations in INDIA :",count_IN)
```

```python
required_stations = {
    "Country" : ["India", "Philippines", "Thailand", "Turkey", "United Kingdom", "United States of America"],
    "Number of Stations" : [count_IN, count_PH, count_TH, count_TR, count_GB, count_US]
}
required_stations_df = pd.DataFrame(required_stations)
required_stations_df
```

```python
required_stations_df["Number of Stations"] = required_stations_df["Number of Stations"].astype("float64")
required_stations_df
```

```python
required_stations_df.dtypes
```

```python
required_boundaries_df = required_boundaries_df.sort_values(by="name")
required_boundaries_df
```

```python
print(required_boundaries_df["geometry"].crs)
```

```python
required_boundaries_df = required_boundaries_df.set_crs(epsg = 4326)
required_boundaries_df = required_boundaries_df.to_crs(epsg = 6933)
required_boundaries_df = required_boundaries_df.reset_index(drop=True)
required_boundaries_df
```

```python
required_stations_df
```

```python
required_stations_df["Area"] = required_boundaries_df["geometry"].area / 1e6
required_stations_df
```

```python
required_stations_df["PM10_density_per_1000sqkm"] = (required_stations_df["Number of Stations"] / (required_boundaries_df["geometry"].area / 1e6)) * 1000
table_df = required_stations_df.sort_values(by="PM10_density_per_1000sqkm", ascending=False)
```
